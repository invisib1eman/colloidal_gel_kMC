```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import json
```

```python
df = pd.read_csv('../../fractal_dimension/fractal_dimension.txt', sep='\s+')
```

```python
df
```

```python
# Filter data for tau_crawl = 1
df_tau1 = df[df['tau_crawl'] == 1.0]

# Create a figure
plt.figure(figsize=(10, 6))

# Get unique salt concentrations
salt_values = df_tau1['salt'].unique()

# Plot Df versus start_index for each salt concentration
for salt in salt_values:
    salt_data = df_tau1[df_tau1['salt'] == salt]
    plt.plot(salt_data['start_index'], salt_data['Df'], marker='o', linestyle='-', label=f'Salt = {salt}')

plt.xlabel('Start Index')
plt.ylabel('Fractal Dimension (Df)')
plt.title('Fractal Dimension vs Start Index for Different Salt Concentrations (Ï„_crawl = 1)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
