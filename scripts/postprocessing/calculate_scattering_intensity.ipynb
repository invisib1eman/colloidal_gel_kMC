```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import re
from IPython.display import display

```

```python
data = pd.read_csv('../../structural_factors/80mM_freeroll_step0.1_N_8000_L_100_t_300_taucrawl_1_start800_end810.csv')
```

```python
display(data)
```

```python
form_factor = np.zeros(len(data))
```

```python
# Particle radius in A
R = 78
```

```python
for k in range(len(data)):
    form_factor[k] = ((4/3) * np.pi * R**3 * (3*(np.sin(data['k_values'][k] * R) - data['k_values'][k] * R * np.cos(data['k_values'][k] * R)) / (data['k_values'][k] * R)**3))**2

plt.plot(data['k_values'], form_factor)
plt.xlabel('k')
plt.ylabel('Form Factor')
plt.title('Form Factor of a Sphere')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
Scattering_intensity = form_factor * data['S_k']
```

```python
plt.plot(data['k_values'], Scattering_intensity)
plt.xlabel('k')
plt.ylabel('Scattering Intensity')
plt.title('Scattering Intensity of a Sphere')
plt.xscale('log')
plt.yscale('log')
plt.xlim(2/(10*R), 2/R)
plt.show()
```

```python
plt.plot(data['k_values'], data['S_k'])
plt.xlabel('k')
plt.ylabel('Structural Factor')
plt.title('Structural Factor of a Sphere')
plt.xscale('log')
plt.yscale('log')
plt.xlim(2/(10*R), 2/R)
plt.show()
```
